```python
from evec_scan.tools import evectools as evt
```

```python
extractor1 = evt.FrameExtractor('evec_scan/datasets/vid1.mp4')
extractor2 = evt.FrameExtractor('evec_scan/datasets/vid2.mp4')
extractor3 = evt.FrameExtractor('evec_scan/datasets/vid3.mp4')
```

```python
fg1 = extractor1.get_random_frame_group(3)
fg2 = extractor2.get_random_frame_group(3)
fg3 = extractor3.get_random_frame_group(3)
```

```python
frames1 = extractor1.get_frames_from_list(fg1)
frames2 = extractor2.get_frames_from_list(fg2)
frames3 = extractor3.get_frames_from_list(fg3)
```

```python
group1 = extractor1.FrameGroup(extractor1, fg1)
group2 = extractor2.FrameGroup(extractor2, fg2)
group3 = extractor3.FrameGroup(extractor3, fg3)
```

```python
for f in fg1:
    group1.show_frame(f)
for f in fg2:
    group2.show_frame(f)
for f in fg3:
    group3.show_frame(f)
```

```python
group3.save_group('vid3Group', 'evec_scan/frame_group_instances')
```
